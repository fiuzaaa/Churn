## Utilização de Random Forest para Identificar Características de Clientes

Nossa análise emprega o algoritmo Random Forest para identificar as características mais relevantes dos clientes.


```python
'''
Módulos para gerar um gráfico de um modelo
'''
from sklearn.tree import export_graphviz
from graphviz import Source
from IPython.display import display, SVG


'''
Exibe uma árvore a partir de um modelo e sua acurácia
'''
def show_tree(model, X, y, target):
  graph_data = export_graphviz(
    model,
    feature_names=columns,
    class_names=target,
    filled=True,
    rounded=True)

  y_pred = model.predict(X)
  print('Acc: %.4f' %accuracy_score(y, y_pred))

  graph = Source(graph_data)
  return SVG(graph.pipe(format='svg'))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import resample
from imblearn.under_sampling import RandomUnderSampler


# desfaz normalizaçao minmax
X_before = scaler.inverse_transform(X)
y = df['cluster'].values

# Objeto RandomUnderSampler
under_sampler = RandomUnderSampler(random_state=42)

# Undersampling aos dados
X_resampled, y_resampled = under_sampler.fit_resample(X_before, y)

# Exiba a distribuição das classes antes e depois do undersampling
print("Distribuição das classes antes do undersampling:")
print(pd.Series(y).value_counts())

print("\nDistribuição das classes após o undersampling:")
print(pd.Series(y_resampled).value_counts())

# Create the Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=0, max_depth=1)

# Train the model on the training data
clf.fit(X_resampled, y_resampled)

# Evaluate on the test data (recommended for unbiased assessment)
accuracy = clf.score(X_resampled, y_resampled)

print('Test Accuracy: %.4f' % accuracy)
```

## Construção de uma Árvore com Condições Identificadas

Imprime uma árvore de decisão que apresenta as condições identificadas para classificar os clientes em grupos distintos.


```python
from treeinterpreter import treeinterpreter as ti
from sklearn.tree import _tree


# Extrair a primeira árvore do modelo
tree = clf[36]

# Obter as regras de decisão da árvore
def extract_decision_rules(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    def recurse(node):
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print(f"if {name} <= {threshold}:")
            recurse(tree_.children_left[node])
            print(f"else:  # if {name} > {threshold}")
            recurse(tree_.children_right[node])
        else:
            valor = tree_.value[node][0]
            print(f"return {round((valor/sum(valor))[1],2)} probabilidade de churn")

    recurse(0)

extract_decision_rules(tree, columns)
```

# Interpretação da Cadeia de Decisões

Esta árvore de decisão parece ser um modelo de classificação para prever a probabilidade de churn (ou seja, a probabilidade de um cliente deixar a empresa). Vamos interpretar cada nó e ramificação:

1. **Se o método de pagamento for Electronic check e for menor ou igual a 0.5:**
   - Se a duração do contrato (tenure) for menor ou igual a 5.5:
     - Se o cliente não tiver serviço de streaming de filmes (StreamingMovies_No internet service) ou for menor ou igual a 0.5, a probabilidade de churn é de 0.81.
     - Se o cliente tiver serviço de streaming de filmes (StreamingMovies_No internet service) e for maior que 0.5, a probabilidade de churn é de 0.44.
   - Se a duração do contrato (tenure) for maior que 5.5:
     - Se o cliente não tiver serviço de segurança online (OnlineSecurity_No internet service) ou for menor ou igual a 0.5, a probabilidade de churn é de 0.35.
     - Se o cliente tiver serviço de segurança online (OnlineSecurity_No internet service) e for maior que 0.5, a probabilidade de churn é de 0.05.

2. **Se o método de pagamento for Electronic check e for maior que 0.5:**
   - Se o contrato for de um ano (Contract_One year) e for menor ou igual a 0.5:
     - Se o cliente não tiver serviço de segurança online (OnlineSecurity_No) ou for menor ou igual a 0.5, a probabilidade de churn é de 0.48.
     - Se o cliente tiver serviço de segurança online (OnlineSecurity_No) e for maior que 0.5, a probabilidade de churn é de 0.81.
   - Se o contrato não for de um ano (Contract_One year) ou for maior que 0.5:
     - Se a cobrança mensal (MonthlyCharges) for menor ou igual a 82.35, a probabilidade de churn é de 0.23.
     - Se a cobrança mensal (MonthlyCharges) for maior que 82.35, a probabilidade de churn é de 0.49.



```python
# Criar uma função para prever com base nas regras da árvore de decisão
def predict(input_features):
    PaymentMethod_Electronic_check = input_features[0]
    tenure = input_features[1]
    StreamingMovies_No_internet_service = input_features[2]
    OnlineSecurity_No_internet_service = input_features[3]
    Contract_One_year = input_features[4]
    OnlineSecurity_No = input_features[5]
    MonthlyCharges = input_features[6]

    if PaymentMethod_Electronic_check <= 0.5:
        if tenure <= 5.5:
            if StreamingMovies_No_internet_service <= 0.5:
                return 1
            else:  # if StreamingMovies_No_internet_service > 0.5
                return 0
        else:  # if tenure > 5.5
            if OnlineSecurity_No_internet_service <= 0.5:
                return 0
            else:  # if OnlineSecurity_No_internet_service > 0.5
                return 0
    else:  # if PaymentMethod_Electronic_check > 0.5
        if Contract_One_year <= 0.5:
            if OnlineSecurity_No <= 0.5:
                return 0
            else:  # if OnlineSecurity_No > 0.5
                return 1
        else:  # if Contract_One_year > 0.5
            if MonthlyCharges <= 82.3499984741211:
                return 0
            else:  # if MonthlyCharges > 82.3499984741211
                return 0

    # if PaymentMethod_Electronic_check <= 0.5:
    #     if tenure <= 5.5:
    #         if StreamingMovies_No_internet_service <= 0.5:
    #             return [58, 243]
    #         else:  # if StreamingMovies_No_internet_service > 0.5
    #             return [68, 53]
    #     else:  # if tenure > 5.5
    #         if OnlineSecurity_No_internet_service <= 0.5:
    #             return [877, 470]
    #         else:  # if OnlineSecurity_No_internet_service > 0.5
    #             return [416, 23]
    # else:  # if PaymentMethod_Electronic_check > 0.5
    #     if Contract_One_year <= 0.5:
    #         if OnlineSecurity_No <= 0.5:
    #             return [133, 121]
    #         else:  # if OnlineSecurity_No > 0.5
    #             return [207, 882]
    #     else:  # if Contract_One_year > 0.5
    #         if MonthlyCharges <= 82.3499984741211:
    #             return [53, 16]
    #         else:  # if MonthlyCharges > 82.3499984741211
    #             return [60, 58]

# Testando a função predict com alguns valores de entrada
input_features = [0.3, 6, 0.2, 0.4, 0.6, 0.3, 70]
prediction = predict(input_features)
print("Previsão:", prediction)
```

```python
# Criar o DataFrame
data = pd.DataFrame(X_before, columns=columns)

# Aplicar a função de previsão a cada linha do DataFrame
predicted_classes = data[['PaymentMethod_Electronic check',
                            'tenure',
                            'StreamingMovies_No internet service',
                            'OnlineSecurity_No internet service',
                            'Contract_One year',
                            'OnlineSecurity_No',
                            'MonthlyCharges']].apply(
    lambda row: predict([row['PaymentMethod_Electronic check'],
                         row['tenure'],
                         row['StreamingMovies_No internet service'],
                         row['OnlineSecurity_No internet service'],
                         row['Contract_One year'],
                         row['OnlineSecurity_No'],
                         row['MonthlyCharges']]),
    axis=1
)

# Adicionar as classes previstas e os rótulos reais ao DataFrame
data['Predicted_Class'] = predicted_classes
data['Churn'] = y

# Exibir apenas as colunas de características originais no DataFrame
features_columns = ['PaymentMethod_Electronic check',
                    'tenure',
                    'StreamingMovies_No internet service',
                    'OnlineSecurity_No internet service',
                    'Contract_One year',
                    'OnlineSecurity_No',
                    'MonthlyCharges',
                    'Predicted_Class',
                    'Churn']
data[features_columns]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns

y_true = data['Churn'].values
y_pred = data['Predicted_Class'].values

# Calcular métricas
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label=1)
recall = recall_score(y_true, y_pred, pos_label=0)
f1 = f1_score(y_true, y_pred, pos_label=1)
conf_matrix = confusion_matrix(y_true, y_pred)

# Exibir métricas
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Confusion Matrix:")
print(conf_matrix)


# Labels para os eixos x e y
labels = ['no Churn', 'Churn']

# Criar o heatmap da matriz de confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g', xticklabels=labels, yticklabels=labels)

# Adicionar rótulos e título
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')

# Mostrar o plot
plt.show()
```
